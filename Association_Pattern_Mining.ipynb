```python
import numpy as np
import pandas as pd
from apyori import apriori

# import data as numpy array
data = np.load("raw_transactions.npy")

# convert numpy array to pandas dataframe
dataset = pd.DataFrame(data)

# get number of records
num_records = len(dataset)
print("Number of records:", num_records, "\n")

# convert dataframe to list of lists 
records = []
for i in range(0, num_records):
    records.append([str(dataset.values[i,j]) for j in range(0, 23)])

# applying apriori
association_rules_3000 = apriori(records, min_support = 3000 / num_records, 
                            min_confidence = 0.8, min_lift = 2, min_length = 2)
# convert rules into list so that it's easier to view results
association_results_3000 = list(association_rules_3000)

# each item corresponds to each rule
print("Length of association results:", len(association_results_3000), "\n")

for i in range(0, len(association_results_3000)):
    print(association_results_3000[i], "\n")
    
# applying apriori
association_rules_5000 = apriori(records, min_support = 5000 / num_records, 
                            min_confidence = 0.8, min_lift = 1, min_length = 2)

# convert rules into list so that it's easier to view results
association_results_5000 = list(association_rules_5000)

print("\n======================================================\n")

# each item corresponds to each rule
print("Length of association results:", len(association_results_5000), "\n")

for i in range(0, len(association_results_5000)):
    print(association_results_5000[i], "\n")

# this prints the results in neater fashion
#=======================================================================
for item in association_results_3000:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    for i in range(len(items)):
        print(items[i], "->")

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")

for item in association_results_5000:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    for i in range(len(items)):
        print(items[i], "->")

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
