```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = np.load('A_data.npy')

A = data[:,[0]]
B = data[:,[1]]

X,Y = data.T

# Problem 1
print("\nMean of X: ", np.mean(A)) 
print("\nMean of Y: ", np.mean(B)) 

print("\nMedian of X: ", np.median(A)) 
print("\nMedian of Y: ", np.median(B)) 

# Problem 2
print("\nSample variance of X: ", np.var(A))
print("\nSample variance of Y: ", np.var(B))

# Problem 3
plt.hist(A, bins = [-10,0,10]) 
plt.title("Histogram for X") 
plt.show()

plt.hist(B, bins = [-20,-10,0,10]) 
plt.title("Histogram for Y") 
plt.show()

# Problem 4
plt.boxplot(A)
plt.title("Boxplot for Y") 
plt.show()

plt.boxplot(B)
plt.title("Boxplot for Y") 
plt.show()

# Problem 5
plt.plot(A[0:3000,0],B[0:3000,0],'x',color = 'blue');
plt.title("Scatterplot for X and Y") 
plt.show()

# Problem 6
xMean = np.mean(A)
yMean = np.mean(B)
print("Vector Mean: \n",[xMean, yMean])

XandY = np.stack((X, Y), axis = 0)
print("\nCovariance Matrix: \n", np.cov(XandY))

# Problem 7
print("\nCorrelation between X and Y: \n", np.corrcoef([X,Y]))

# Problem 8
dX = pd.cut(X, 5)
dY = pd.cut(Y, 5)

print("\ndX is: \n", dX)
print("\ndY is: \n", dY)

# Problem 9
plt.hist(A, density = True, bins = 5)
plt.title("Histogram for Discretized X") 
plt.show()

plt.hist(B, density = True, bins = 5);
plt.title("Histogram for Discretized Y") 
plt.show()

# Problem 10
df = pd.DataFrame(dX, dY)
print ("\nContingency Table: \n", df)

# Create heatmap
heatmap, xedges, yedges = np.histogram2d(X, Y, bins = (5,5))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

# Plot heatmap
plt.clf()
plt.title('Heatmap for Discretized Data')
plt.imshow(heatmap, extent = extent)
plt.show()
```
